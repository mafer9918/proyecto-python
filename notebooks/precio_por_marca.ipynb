```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os

#Crear directorio
os.makedirs("../assets/", exist_ok=True)

#Leer data
data_path = "../data/processed/clean_products.csv"
df = pd.read_csv(data_path)

#Generar grfica
df['brand'] = df['brand'].apply(lambda x: x.split()[0])
price_by_brand = df.groupby('brand')['price'].mean().sort_values()
price_by_brand_reset = price_by_brand.reset_index()
plt.figure(figsize=(12, 8))
sns.barplot(data=price_by_brand_reset, x='brand', y='price')
plt.title('Precio Medio por Marca')
plt.xlabel('Marca')
plt.ylabel('Precio Medio')
plt.xticks(rotation=45)
plt.savefig(os.path.join("../assets/", 'precio_por_marca.png'))
plt.show()
plt.close()
```
