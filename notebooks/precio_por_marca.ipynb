```python
import pandas as pd # Importamos la librera pandas para la manipulacin de la data.
import matplotlib.pyplot as plt # Importamos la libreria matplotlib para visualizacion de datos.
import seaborn as sns # Importamos la libreria seaborn para la visualizacion de datos.
import os # Importamos os para el manejo del sistema de archivos

#Crear directorio para almacenar la imagen generada
os.makedirs("../assets/", exist_ok=True)

#Cargamos los datos procesados.
data_path = "../data/processed/clean_products.csv"
df = pd.read_csv(data_path)

# Anlisis y visualizacin
df['brand'] = df['brand'].apply(lambda x: x.split()[0]) #Tomar la primera palabra de cada marca
price_by_brand = df.groupby('brand')['price'].mean().sort_values().reset_index() #Calcular el precio promedio por marca y ordernamos.
plt.figure(figsize=(12, 8)) #Configurar el tamao de la figura
sns.barplot(data=price_by_brand, x='brand', y='price') # Creamos un grfico de barras con seaborn.
plt.title('Precio Medio por Marca') #Titulo del grfico
plt.xlabel('Marca') #Agregar etiqueta al eje X
plt.ylabel('Precio Medio') #Agregar etiqueta al eje Y
plt.xticks(rotation=45) # Rotamos las etiquetas del eje x para mayor legibilidad.
plt.tight_layout() # Ajustamos el diseo de la figura.
plt.savefig(os.path.join("../assets/", 'precio_por_marca.png')) # Guardamos la figura en el directorio "assets".
plt.show() # Mostrar el grfico de barras
```
